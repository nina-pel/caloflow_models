```python
import numpy as np  
import h5py
import matplotlib.pyplot as plt
import torch
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import accuracy_score
from nflows.flows.base import Flow
from nflows.distributions.normal import StandardNormal
from nflows.transforms.base import CompositeTransform
from nflows.transforms.coupling import PiecewiseRationalQuadraticCouplingTransform 
from nflows.transforms.permutations import ReversePermutation
from nflows.transforms.coupling import AffineCouplingTransform
import time
import torch.nn as nn
import torch.optim as optim
import os

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

ALPHA = 1e-6

results_outdir = "CLNF_affine_results"
os.makedirs(results_outdir, exist_ok=True)
results_txt_path = os.path.join(results_outdir, "clnf_affine_results.txt")
results_log = []

def log_print(msg=""):
    print(msg)
    results_log.append(str(msg))

def format_runtime(seconds):
    hours, rem = divmod(seconds, 3600)
    minutes, seconds = divmod(rem, 60)
    return f"{int(hours)}h {int(minutes)}m {seconds:.2f}s"

def save_results_txt():
    with open(results_txt_path, "w") as f:
        f.write("\n".join(results_log))

def logit(x):
    return torch.log(x / (1.0 - x))

def logit_trafo(x):
    local_x = torch.clamp(ALPHA + (1. - 2. * ALPHA) * x, min=1e-6, max=1 - 1e-6)
    return logit(local_x)

def inverse_logit(x):
    return (torch.sigmoid(x) - ALPHA) / (1. - 2.*ALPHA)

part1_start = time.time()

path_to_data = '/groups/hephy/mlearning/CaloRegression/training/'
my_file = h5py.File(path_to_data + 'train_piplus.hdf5', 'r')
energy = my_file['energy'][:]
layers = [my_file[f'layer_{i}'][:] for i in range(6)]
E_layers = np.array([layer.sum(axis=(1, 2)) for layer in layers]).T

data = np.hstack((energy.reshape(-1, 1), E_layers))
data[np.isinf(data)] = np.nan
valid_mask = ~np.isnan(data).any(axis=1)
data = data[valid_mask]

data_tensor = torch.tensor(data, dtype=torch.float32).to(device)

data_max = data_tensor.max(dim=0).values
data_tensor = data_tensor / data_max
data_tensor = logit_trafo(data_tensor)

mask = ~torch.isnan(data_tensor).any(dim=1) & ~torch.isinf(data_tensor).any(dim=1)
data_tensor = data_tensor[mask]

data_mean = data_tensor.mean(dim=0)
data_std = data_tensor.std(dim=0)
data_tensor = (data_tensor - data_mean) / data_std

data_mean_reg = data_mean.clone()
data_std_reg = data_std.clone()
data_max_reg = data_max.clone()

context_feature = data_tensor[:, 0].unsqueeze(1)
input_features = data_tensor[:, 1:]

num_layers = 30  
num_features = input_features.shape[1]
base_distribution = StandardNormal(shape=[num_features])

transforms = []

masks = [(torch.arange(num_features) % 2).bool() for _ in range(num_layers)]
masks = [mask if i % 2 == 0 else ~mask for i, mask in enumerate(masks)]

masks = []
for i in range(num_layers):
    if i % 2 == 0:
        masks.append(torch.tensor(np.random.choice([0,0,0,1,1,1], replace=False,size=6)))
    else:
        masks.append(1-masks[-1])

print(masks)

class CouplingMLP(nn.Module):
    def __init__(self, in_features, out_features, context_features=1):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_features + context_features, 128), nn.LeakyReLU(0.01),
            nn.Linear(128, 128), nn.LeakyReLU(0.01),
            nn.Linear(128, out_features)
        )

    def forward(self, identity_split, context):
        inputs = torch.cat([identity_split, context], dim=-1)
        return self.net(inputs)

for i in range(num_layers):
    transforms.append(AffineCouplingTransform(
        mask=masks[i],
        transform_net_create_fn=lambda in_features, out_features: CouplingMLP(in_features, out_features, context_features=1)
    ))

transform = CompositeTransform(transforms)
flow = Flow(transform, base_distribution).to(device)

def init_weights(m):
    if isinstance(m, nn.Linear):
        nn.init.xavier_uniform_(m.weight)
        if m.bias is not None:
            nn.init.zeros_(m.bias)

flow.apply(init_weights)

optimizer = torch.optim.Adam(flow.parameters(), lr=1e-3)
num_epochs = 500
batch_size = 256
best_loss = float('inf')

for epoch in range(num_epochs):
    idx = torch.randperm(len(data_tensor))
    data_tensor = data_tensor[idx]
    context_feature = data_tensor[:, 0].unsqueeze(1)
    input_features = data_tensor[:, 1:]

    for i in range(0, len(input_features), batch_size):
        batch_inputs = input_features[i:i + batch_size]
        batch_context = context_feature[i:i + batch_size]
        optimizer.zero_grad()
        loss = -flow.log_prob(inputs=batch_inputs, context=batch_context).mean()
        if not torch.isnan(loss) and not torch.isinf(loss):
            loss.backward()
            optimizer.step()
            best_loss = min(best_loss, loss.item())
        else:
            print("skipping")
    
    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch + 1}, Loss: {loss.item()}")
        
n_samples = 200000
batch_size = 1000
all_samples = []

Einc_gen_phys = torch.rand((n_samples, 1), device=device) * 99 + 1
Einc_gen = Einc_gen_phys / data_max[0]
Einc_gen = logit_trafo(Einc_gen)
Einc_gen = (Einc_gen - data_mean[0]) / data_std[0]

with torch.no_grad():
    for i in range(0, n_samples, batch_size):
        batch_context = Einc_gen[i:i + batch_size]
        samples_batch = flow.sample(1, context=batch_context).detach().cpu()
        all_samples.append(samples_batch)

samples = torch.cat(all_samples, dim=0)
samples = samples * data_std[1:].cpu() + data_mean[1:].cpu()
samples = inverse_logit(samples)
samples = samples * data_max[1:].cpu()
samples = samples.squeeze().numpy()

fig, axs = plt.subplots(2, 3, figsize=(18, 12))

axs = axs.flatten()
colors = ['yellow', 'green', 'red', 'purple', 'orange', 'cyan']

for i in range(data.shape[1] - 1):
    bins = np.logspace(0, 5, 100)
    counts, bin_edges, _ = axs[i].hist(
        data[:, i + 1], bins=bins, alpha=0.5, label="Reference Data", color=colors[i % len(colors)], density=True, log=True
    )
    axs[i].hist(samples[:, i], bins=bin_edges, alpha=0.5, label="Generated Data", color='blue', density=True, log=True)
    axs[i].set_xscale('log')
    axs[i].legend()
    axs[i].set_title(f"Feature {i + 1}")

for j in range(len(axs)):
    if j >= data.shape[1] - 1:
        axs[j].axis('off')

plt.tight_layout()
outpath = os.path.join(results_outdir, "CLNF_layer_energy_histograms_new.png")
plt.savefig(outpath, dpi=600, bbox_inches="tight")
log_print(f"Saved: {outpath}")
plt.show()

part1_time = time.time() - part1_start

from sklearn.calibration import calibration_curve
from sklearn.isotonic import IsotonicRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import copy

part2_start = time.time()

ALPHA = 1e-6

real_data = data.copy()
generated_data = np.hstack((Einc_gen_phys.detach().cpu().numpy(), samples))

real_data = torch.tensor(real_data, dtype=torch.float32)
generated_data = torch.tensor(generated_data, dtype=torch.float32)

real_data = real_data / data_max_reg.detach().cpu()
generated_data = generated_data / data_max_reg.detach().cpu()

real_data = logit_trafo(real_data)
generated_data = logit_trafo(generated_data)

real_data = (real_data - data_mean_reg.detach().cpu()) / data_std_reg.detach().cpu()
generated_data = (generated_data - data_mean_reg.detach().cpu()) / data_std_reg.detach().cpu()

real_data = real_data.numpy()
generated_data = generated_data.numpy()

real_data = real_data[:, 1:]
generated_data = generated_data[:, 1:]

smallest_size = min(len(real_data), len(generated_data))

real_indices = np.random.choice(len(real_data), smallest_size, replace=False)
generated_indices = np.random.choice(len(generated_data), smallest_size, replace=False)

real_data = real_data[real_indices]
generated_data = generated_data[generated_indices]

print(f"Final real data shape: {real_data.shape}")
print(f"Final gen shape: {generated_data.shape}")

real_labels = np.ones(len(real_data))  
maf_labels = np.zeros(len(generated_data))

combined_data = np.vstack((real_data, generated_data))
combined_labels = np.hstack((real_labels, maf_labels))

X_train, X_test, y_train, y_test = train_test_split(combined_data, combined_labels, test_size=0.2, random_state=42)

class SimpleNN(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, 1)
        self.dropout = nn.Dropout(0.3)
        
    def forward(self, x):
        x = F.leaky_relu(self.fc1(x), negative_slope=0.01)
        x = self.dropout(x)
        x = F.leaky_relu(self.fc2(x), negative_slope=0.01)
        x = self.dropout(x)
        x = torch.sigmoid(self.fc3(x))
        return x

def train_model(model, criterion, optimizer, dataloaders, num_epochs=25):
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0
    
    for epoch in range(num_epochs):
        print('Epoch number {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device).float().unsqueeze(1)
                
                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                    preds = torch.round(outputs)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / len(dataloaders[phase].dataset)
            epoch_acc = running_corrects.double() / len(dataloaders[phase].dataset)

            print('{} Loss: {:.9f} Acc: {:.9f}'.format(
                phase, epoch_loss, epoch_acc))

            if phase == 'val' and epoch_acc  > best_acc:
                    best_acc = epoch_acc
                    best_model_wts = copy.deepcopy(model.state_dict())

    model.load_state_dict(best_model_wts)
    return model, float(best_acc)

def evaluate_model(model, dataloaders):
    model.eval()
    all_preds = []
    all_labels = []
    all_preds_auc = []

    for inputs, labels in dataloaders['val']:
        inputs = inputs.to(device)
        labels = labels.to(device).float().unsqueeze(1)
        
        with torch.no_grad():
            outputs = model(inputs)
            preds = torch.round(outputs)
            preds_auc = outputs
            all_preds_auc.append(preds_auc.cpu().numpy())
            all_preds.append(preds.cpu().numpy())
            all_labels.append(labels.cpu().numpy())
    
    all_preds_auc = np.concatenate(all_preds_auc)
    all_preds = np.concatenate(all_preds)
    all_labels = np.concatenate(all_labels)
    acc = accuracy_score(all_labels, all_preds)
    auc = roc_auc_score(all_labels, all_preds_auc)
    return acc, auc

train_dataset = torch.utils.data.TensorDataset(torch.tensor(X_train, dtype=torch.float32), torch.tensor(y_train, dtype=torch.float32))
test_dataset = torch.utils.data.TensorDataset(torch.tensor(X_test, dtype=torch.float32), torch.tensor(y_test, dtype=torch.float32))

dataloaders = {
    'train': torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True),
    'val': torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=True)
}

input_dim = combined_data.shape[1]
hidden_dim = 64
model = SimpleNN(input_dim, hidden_dim)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = model.to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

model, best_val_acc = train_model(model, criterion, optimizer, dataloaders, num_epochs=25)
final_acc, final_auc = evaluate_model(model, dataloaders)

part2_time = time.time() - part2_start

part3_start = time.time()

energy_values = list(range(10, 91, 10)) 
eval_folder = '/groups/hephy/mlearning/CaloRegression/evaluation'
errors_by_energy = {}
print(data_max_reg[0])

SAVE_DPI = 400
SAVE_FORMAT = "pdf"

def logit_reg(x):
    return torch.log(x / (1.0 - x))

def logit_trafo_reg(x):
    local_x = torch.clamp(ALPHA + (1. - 2. * ALPHA) * x, min=1e-6, max=1 - 1e-6)
    return logit_reg(local_x)

def inverse_logit_reg(x):
    return (torch.sigmoid(x) - ALPHA) / (1. - 2.*ALPHA)

def load_and_preprocess_test_file(filepath, data_mean_reg, data_std_reg, data_max_reg):
    with h5py.File(filepath, 'r') as f:
        energy = f['energy'][:]
        layers = [f[f'layer_{i}'][:] for i in range(6)]
        E_layers = np.array([layer.sum(axis=(1, 2)) for layer in layers]).T

    data = np.hstack((energy.reshape(-1, 1), E_layers))
    data[np.isinf(data)] = np.nan
    valid_mask = ~np.isnan(data).any(axis=1)
    data = data[valid_mask]

    tensor_data = torch.tensor(data, dtype=torch.float32).to(device)
    tensor_data = tensor_data / data_max_reg  
    tensor_data = logit_trafo_reg(tensor_data)  
    tensor_data = (tensor_data - data_mean_reg) / data_std_reg 

    return tensor_data[:, 1:]

def infer_incident_energy(flow, layer_tensor, data_mean_reg, data_std_reg, data_max_reg, num_scan_points=1000):
    device = layer_tensor.device
    n_samples = layer_tensor.shape[0]
    max_energy = float(data_max_reg[0].item())
    E_candidates = torch.linspace(1, max_energy, num_scan_points, device=device).view(-1, 1)

    E_candidates_norm = E_candidates / data_max_reg[0]
    E_candidates_logit = logit_trafo_reg(E_candidates_norm)
    E_candidates_std = (E_candidates_logit - data_mean_reg[0]) / data_std_reg[0]

    E_context = E_candidates_std.view(1, -1, 1).repeat(n_samples, 1, 1)
    inputs = layer_tensor.unsqueeze(1).repeat(1, num_scan_points, 1)

    flat_inputs = inputs.view(-1, inputs.shape[-1])
    flat_context = E_context.view(-1, 1)

    batch_size = 1000  
    log_probs_list = []
    with torch.no_grad():
        for i in range(0, flat_inputs.shape[0], batch_size):
            inputs_batch = flat_inputs[i:i + batch_size]
            context_batch = flat_context[i:i + batch_size]
            log_probs_batch = flow.log_prob(inputs_batch, context=context_batch)
            log_probs_list.append(log_probs_batch)

    log_probs = torch.cat(log_probs_list).view(n_samples, num_scan_points)
    max_indices = log_probs.argmax(dim=1)
    inferred_E = E_candidates[max_indices].cpu().numpy()

    return inferred_E

data_mean_reg = data_mean_reg.to(device)
data_std_reg = data_std_reg.to(device)    
data_max_reg = data_max_reg.to(device)   
flow = flow.to(device)   

errors_by_energy = {}

for E_true in energy_values:
    file_path = os.path.join(eval_folder, f'piplus_{E_true}GeV.hdf5')

    layer_tensor = load_and_preprocess_test_file(
        file_path, data_mean_reg, data_std_reg, data_max_reg
    )
    inferred = infer_incident_energy(
        flow, layer_tensor, data_mean_reg, data_std_reg, data_max_reg
    )

    error = inferred - E_true
    errors_by_energy[E_true] = error

    log_print(f"E_inc={E_true} GeV: mean error = {np.mean(error):.2f}, std = {np.std(error):.2f}")

    bins = 100
    counts, bin_edges = np.histogram(inferred, bins=bins, range=(1, float(data_max_reg[0].item())))

    bin_idx = np.digitize(E_true, bin_edges) - 1
    bin_idx = np.clip(bin_idx, 0, len(counts) - 1)

    true_bin_count = counts[bin_idx]
    ratio_true_bin = true_bin_count / counts.sum() if counts.sum() > 0 else 0
    log_print(f"  Bin count of true energy ({E_true} GeV): {true_bin_count} ({ratio_true_bin:.2%})")

    max_count = counts.max()
    half_max = max_count / 2
    above_half_max = np.where(counts >= half_max)[0]
    if len(above_half_max) > 0:
        fwhm = bin_edges[above_half_max[-1] + 1] - bin_edges[above_half_max[0]]
    else:
        fwhm = 0.0

    log_print(f"  Full width at half maximum (FWHM): {fwhm:.2f} GeV")

plt.figure(figsize=(12, 6))

for i, (E_true, error) in enumerate(errors_by_energy.items()):
    inferred = error + E_true
    plt.hist(inferred, bins=100, range=(0, 100), alpha=0.5, label=f'True {E_true} GeV')
    plt.axvline(E_true, color='black', linestyle='--')

plt.xlabel("Inferred incident energy (GeV)")
plt.ylabel("Counts")
plt.title("Inferred energy distributions by true $E_{inc}$ (CLNF - affine)")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig(
    os.path.join(results_outdir, "AffineCLNF_inferred_energy_distributions_new.pdf"),
    dpi=SAVE_DPI,
    bbox_inches="tight"
)
plt.show()

plt.figure(figsize=(12, 6))
for E_true, error in errors_by_energy.items():
    plt.hist(error, bins=50, alpha=0.5, label=f'{E_true} GeV')

plt.axvline(0, color='black', linestyle='--')
plt.xlabel("E_found - E_inc (GeV)")
plt.ylabel("Counts")
plt.title("Error distribution for different true energies (CLNF - affine)")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig(
    os.path.join(results_outdir, "AffineCLNF_error_distributions_by_energy_new.pdf"),
    dpi=SAVE_DPI,
    bbox_inches="tight"
)
plt.show()

test_file_path = '/groups/hephy/mlearning/CaloRegression/evaluation/piplus_20GeV.hdf5'

test_layers_tensor = load_and_preprocess_test_file(test_file_path, data_mean_reg, data_std_reg, data_max_reg)
print("=== Test Input Diagnostics ===")
print("test_layers_tensor[:5]:", test_layers_tensor[:5].cpu())
print("test_layers_tensor.std():", test_layers_tensor.std().item())

max_energy = float(data_max_reg[0].item())
E_candidates = torch.linspace(1, max_energy, 1000, device=device).view(-1, 1)
E_candidates_norm = E_candidates / data_max_reg[0]
E_candidates_logit = logit_trafo_reg(E_candidates_norm)
E_candidates_std = (E_candidates_logit - data_mean_reg[0]) / data_std_reg[0]

print("\n=== Debug Info ===")
print("data_max_reg[0]:", data_max_reg[0].item())
print("data_mean_reg[0]:", data_mean_reg[0].item())
print("data_std_reg[0]:", data_std_reg[0].item())
print("E_candidates[:5]:", E_candidates[:5].view(-1).cpu().numpy())
print("E_candidates_norm[:5]:", E_candidates_norm[:5].view(-1).cpu().numpy())
print("E_candidates_logit[:5]:", E_candidates_logit[:5].view(-1).cpu().numpy())
print("E_candidates_std[:5]:", E_candidates_std[:5].view(-1).cpu().numpy())

def infer_incident_energy_debug(flow, layer_tensor, data_mean_reg, data_std_reg, data_max_reg, num_scan_points=1000):
    device = layer_tensor.device
    n_samples = layer_tensor.shape[0]

    max_energy = float(data_max_reg[0].item())
    E_candidates = torch.linspace(1, max_energy, num_scan_points, device=device).view(-1, 1)

    E_candidates_norm = E_candidates / data_max_reg[0]
    E_candidates_logit = logit_trafo_reg(E_candidates_norm)
    E_candidates_std = (E_candidates_logit - data_mean_reg[0]) / data_std_reg[0]

    E_context = E_candidates_std.view(1, -1, 1).repeat(n_samples, 1, 1)
    inputs = layer_tensor.unsqueeze(1).repeat(1, num_scan_points, 1)

    flat_inputs = inputs.view(-1, inputs.shape[-1])
    flat_context = E_context.view(-1, 1)
    print("\n=== Debug Info ===")
    print("data_max_reg[0]:", data_max_reg[0].item())
    print("data_mean_reg[0]:", data_mean_reg[0].item())
    print("data_std_reg[0]:", data_std_reg[0].item())
    print("E_candidates[:5]:", E_candidates[:5].view(-1).cpu().numpy())
    print("E_candidates_norm[:5]:", E_candidates_norm[:5].view(-1).cpu().numpy())
    print("E_candidates_logit[:5]:", E_candidates_logit[:5].view(-1).cpu().numpy())
    print("E_candidates_std[:5]:", E_candidates_std[:5].view(-1).cpu().numpy())

    batch_size = 10000
    log_probs_list = []
    with torch.no_grad():
        for i in range(0, flat_inputs.shape[0], batch_size):
            inputs_batch = flat_inputs[i:i + batch_size]
            context_batch = flat_context[i:i + batch_size]
            log_probs_batch = flow.log_prob(inputs_batch, context=context_batch)
            log_probs_list.append(log_probs_batch)

    log_probs = torch.cat(log_probs_list).view(n_samples, num_scan_points)
    print("\n=== Log Prob Diagnostics ===")
    print("log_probs[0][:10]:", log_probs[0][:10].cpu().numpy())
    print("log_probs[0].max():", log_probs[0].max().item())
    print("log_probs[0].min():", log_probs[0].min().item())
    print("log_probs[0].std():", log_probs[0].std().item())

    max_indices = log_probs.argmax(dim=1)
    inferred_E = E_candidates[max_indices].cpu().numpy()

    return inferred_E

inferred_energies = infer_incident_energy_debug(flow, test_layers_tensor, data_mean_reg, data_std_reg, data_max_reg)
random_indices = torch.randperm(len(inferred_energies))[:20]
print(inferred_energies[random_indices])

plt.figure(figsize=(10, 6))
plt.hist(inferred_energies, bins=100, range=(0, 100), alpha=0.7, color='skyblue', label="Inferred $E_{inc}$")
plt.axvline(20, color='black', linestyle='--', label='True 20 GeV')
plt.xlabel("Inferred Incident Energy (GeV)")
plt.ylabel("Counts")
plt.legend()
plt.title("Regressive Evaluation of Flow Model (E_inc = 20 GeV) (CLNF - affine)")
plt.grid(True)
plt.tight_layout()
plt.savefig(
    os.path.join(results_outdir, "AffineCLNF_regressive_evaluation_20GeV_new.pdf"),
    dpi=SAVE_DPI,
    bbox_inches="tight"
)
plt.show()

import numpy as np

def width68(errors):
    if len(errors) == 0:
        return np.nan
    return np.quantile(errors, 0.84) - np.quantile(errors, 0.16)

def metrics(errors):
    errors = np.asarray(errors).reshape(-1)
    if errors.size == 0:
        return 0, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan
    bias = float(np.mean(errors))
    mae  = float(np.mean(np.abs(errors)))
    rmse = float(np.sqrt(np.mean(errors**2)))
    std  = float(np.std(errors))
    median = float(np.median(errors))
    w68  = float(width68(errors))
    return int(errors.size), bias, mae, rmse, std, median, w68

def print_regression_summary(errors_by_energy, sat_by_energy=None, filter_saturated=True):
    energies = sorted(errors_by_energy.keys())

    pooled_errors = []
    pooled_trueE = []

    tag = "edge-saturation filtered" if filter_saturated else "unfiltered"
    log_print("")
    log_print(f"Per-energy regression summary ({tag})")
    log_print("-------------------------------------------------------")
    log_print(f"{'E_true':>6} {'N':>10} {'Bias':>10} {'MAE':>10} {'RMSE':>10} {'Std':>10} {'Median':>10} {'68% width':>12} {'Sat.frac':>10}")

    for E in energies:
        err = np.asarray(errors_by_energy[E]).reshape(-1)

        if sat_by_energy is None:
            sat = np.zeros(err.size, dtype=bool)
        else:
            sat = np.asarray(sat_by_energy[E], dtype=bool).reshape(-1)
            if sat.size != err.size:
                raise ValueError(f"sat_by_energy[{E}] has size {sat.size}, expected {err.size}")

        sat_frac = (np.mean(sat) * 100.0) if err.size > 0 else np.nan

        use = err[~sat] if filter_saturated else err

        N, bias, mae, rmse, std, median, w68 = metrics(use)

        log_print(f"{int(E):6d} {N:10d} {bias:10.3f} {mae:10.3f} {rmse:10.3f} {std:10.3f} {median:10.3f} {w68:12.3f} {sat_frac:9.2f}%")

        pooled_errors.append(use)
        pooled_trueE.append(np.full(use.size, float(E)))

    pooled_errors = np.concatenate(pooled_errors) if sum(map(len, pooled_errors)) > 0 else np.array([])
    pooled_trueE  = np.concatenate(pooled_trueE)  if sum(map(len, pooled_trueE)) > 0 else np.array([])

    N, bias, mae, rmse, std, median, w68 = metrics(pooled_errors)

    rel_mae  = float(np.mean(np.abs(pooled_errors) / pooled_trueE) * 100.0) if N > 0 else np.nan
    rel_rmse = float(np.sqrt(np.mean((pooled_errors / pooled_trueE) ** 2)) * 100.0) if N > 0 else np.nan

    log_print("")
    log_print("Global regression summary (all energies pooled)")
    log_print("----------------------------------------------")
    log_print(f"Total N: {N}")
    log_print(f"Bias [GeV]: {bias:.3f}")
    log_print(f"MAE [GeV]: {mae:.3f}")
    log_print(f"RMSE [GeV]: {rmse:.3f}")
    log_print(f"Std [GeV]: {std:.3f}")
    log_print(f"68% width [GeV]: {w68:.3f}")
    log_print(f"Relative MAE [%]: {rel_mae:.3f}")
    log_print(f"Relative RMSE [%]: {rel_rmse:.3f}")

part3_time = time.time() - part3_start

log_print("CLNF - affine results")
log_print("====================")
log_print(f"Final best loss: {best_loss:.6f}")
log_print(f"Final Accuracy: {final_acc:.6f}")
log_print(f"Final AUC: {final_auc:.6f}")
log_print("")
log_print("Timing")
log_print("------")
log_print(f"1. Generative part without classifier: {format_runtime(part1_time)}")
log_print(f"2. Classifier: {format_runtime(part2_time)}")
log_print(f"3. Regression: {format_runtime(part3_time)}")
log_print(f"Total time taken: {format_runtime(part1_time + part2_time + part3_time)}")

print_regression_summary(errors_by_energy, sat_by_energy=None, filter_saturated=True)

save_results_txt()
```
